# SmartClaim AI - Dataset Exploration

This notebook explores the Saudi car accident report dataset, analyzes text distributions, and visualizes embeddings.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from src.preprocess import load_data

# Load the dataset
df = load_data('../data/smartclaim_expanded_dataset.xlsx')
df.head()
```

## 1. Cost Distribution Analysis

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['cost'], kde=True, bins=30)
plt.title("Repair Cost Distribution (SAR)")
plt.xlabel("Cost (SAR)")
plt.ylabel("Frequency")
plt.show()
```

## 2. Text Length Analysis

```python
df['text_length'] = df['text'].apply(len)
df['word_count'] = df['text'].apply(lambda x: len(x.split()))

plt.figure(figsize=(10, 6))
sns.histplot(df['word_count'], bins=20, color='orange')
plt.title("Word Count Distribution in Reports")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

## 3. Sample Arabic Texts

```python
print("Sample Accident Reports:")
for i, text in enumerate(df['text'].sample(5)):
    print(f"{i+1}. {text}")
```

## 4. Embedding Visualization (PCA)

We use PCA to reduce the 768D embeddings to 2D for visualization.

```python
from sklearn.decomposition import PCA
from src.embed import get_embeddings

sample_df = df.sample(200)
embeddings = get_embeddings(sample_df['text'].tolist())

pca = PCA(n_components=2)
pca_results = pca.fit_transform(embeddings)

plt.figure(figsize=(12, 8))
scatter = plt.scatter(pca_results[:, 0], pca_results[:, 1], c=sample_df['cost'], cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='Cost (SAR)')
plt.title("PCA Visualization of Arabic Accident Report Embeddings")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

## 5. Correlation Analysis

Is there a simple correlation between text length and cost?

```python
correlation = df[['cost', 'text_length', 'word_count']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
